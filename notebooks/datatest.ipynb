```python
from pathlib import Path
import os
import glob
import shutil
import cv2
import matplotlib.pyplot as plt

BASE_DIR = Path("..").resolve()
RAW_DIR = BASE_DIR / "data" / "SurgiSR4K"
IMG_ROOT = RAW_DIR / "data" / "images" / "480x270p"

print("Image root:", IMG_ROOT)
print("Subfolders:")
for p in IMG_ROOT.iterdir():
    if p.is_dir():
        print("  -", p.name)
```

```python
COLLECT_DIR = BASE_DIR / "data" / "processed" / "all_frames_480p"
COLLECT_DIR.mkdir(parents=True, exist_ok=True)

all_pngs = glob.glob(str(IMG_ROOT / "*" / "*.png"))
print("Found", len(all_pngs), "PNG frames in all subfolders.")
```

```python
MAX_IMAGES = 600

for idx, src_path in enumerate(all_pngs[:MAX_IMAGES]):
    fname = f"frame_{idx:04d}.png"
    dst_path = COLLECT_DIR / fname
    shutil.copy(src_path, dst_path)

print("Copied", min(len(all_pngs), MAX_IMAGES), "frames into:", COLLECT_DIR)
```

```python
import random

sample = random.choice(os.listdir(COLLECT_DIR))
img = cv2.imread(str(COLLECT_DIR / sample))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))
plt.imshow(img)

plt.axis("off")
plt.show()
```
